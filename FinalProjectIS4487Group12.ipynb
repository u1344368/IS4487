**Business Challenge**

Frontier Airlines has a major problem with the number of flight cancellations they have per year. In 2024, they were ranked with the highest number of flights cancelled at 2.7 % compared to Delta, who had the lowest percentage of flights cancelled at 0.2%. This starts to cause major issues for Frontier. It affects customer trust, causes damage to the brand, adds financial costs with rebooking, and there is an operational inefficiency as well as competitor risk. Frequent cancellations not only upsets customers, causing a bad reputation for the brand, but it is also costing Frontier thousands of dollars.

At the beginning of this year, the U.S. Department of Transportation fined Frontier Airlines $650,000 in civil penalties for operating multiple chronically delayed flights. This has a huge impact on the company as a whole. If Frontier had a way they could predict when flights would be delayed, then they would be able to take action against this and therefore avoid big fines as well as improve customer satisfaction.
